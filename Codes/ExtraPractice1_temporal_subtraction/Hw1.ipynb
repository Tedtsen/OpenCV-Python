```python
#temporal subtraction simple

import cv2
cap = cv2.VideoCapture(0)
# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")
cur_frame, prev_frame = None, None
while True:
    ret, frame = cap.read()
    prev_frame = cur_frame
    cur_frame = cv2.resize(frame, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)
    if prev_frame is not None:
        sub_frame = cv2.absdiff(cur_frame , prev_frame)
        cv2.imshow('temporal subtraction', sub_frame)
    c = cv2.waitKey(1)
    if c == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#motion history image

#Code for Homework 1
import cv2
alpha = 0.8
cap = cv2.VideoCapture(0)
# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")
    
ret, frame = cap.read()
cur_frame = mhi_frame = cv2.resize(frame, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_AREA)
while True:
    ret, frame = cap.read()
    cur_frame = cv2.resize(frame, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_AREA)
    mhi_frame = cv2.addWeighted(mhi_frame, alpha, cur_frame, 1-alpha, 0)
    backsub_frame = cv2.absdiff(cur_frame, mhi_frame)
    cv2.imshow('background subtraction image', backsub_frame)
    cv2.imshow('background image (dynamic background model)', mhi_frame)
    c = cv2.waitKey(1)
    if c == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#5. Play around the forgetting factor 0<α<1. Discuss the effect.
    B(0)=I(0),
    B(t)=αB(t-1)+(1- α)I(t)
    From the equation, we can see that it is a recursive function
    where for every time unit after 0, the output image B(t) is constructed
    from the previous output image B(t-1) with a weight a
    and the current input image I(t) with a weight of 1-a.
    
    #For background image
    When we set the forgetting factor to 0, the previous output image B(t-1)
    will be discarded as it multiplies by 0 and output image will consist 
    of only the current input image as it multiplies by 1, 
    so the output image is the same as current input. B(t) = I(t).
    
    When we set the forgetting factor to 1, the previous output image B(t-1)
    will be the output image and the current input image is discarded. We
    can only see the first input image as the consequent input images after 
    t=0 are discarded as they all multiply by 0.
    B(t) = B(t-1) = B(0) = I(0).
    
    For any forgetting factor in between and excluding 0 and 1, the higher we 
    set it, the slower the output image will forget the previous background 
    image, so we can see the motion history effect. When the forgetting factor
    is high, the previous frame fades out slower, preserving the image of the moving
    object old location longer and making the newer frame fades in slower. Vice
    versa when forgetting factor is lower.
    
    #For subtraction image
    When forgetting factor is 0. Background image is current input frame.
    B(t) = I(t). When doing background subtraction, we do abs(I(t)-B(t)),
    so the equation becomes abs(I(t)-I(t)), therefore output image is black
    and blank.
    
    When forgetting factor is 1. Background image is I(0), or the first input
    frame, as the equation goes like B(t) = B(t-1) = B(0) = I(0). Therefore,
    when doing background subtraction abs(I(t)-B(t)), the equation will become
    abs(I(t)-I(0)), so the output image will be the highlight of the objects
    that were not present in the first input image.
    
    When the forgetting factor is in between and excluding 0 and 1. The
    higher we set the forgetting factor, the more B(t) will contain traces of frames
    from previous time units, so the background subtraction equation 
    abs(I(t)-B(t)) will become similar to abs(I(t)-I(t-1)), so we get the effect
    similar to that of temporal subtraction, except that there is also a trail of the
    moving object.
    
#6. Compare the difference between the temporal subtraction and 
#   background subtraction. Write down observation.
    Temporal subtraction is where we do abs(I(t)-I(t-1)) while background subtraction
    is where we do abs(I(t)-B(t)). 
    Temporal subtraction seems that constant moving objects are highlighted in the output
    frame, when their positions in the images are different in I(t) and I(t-1). If everything
    is the same and static with their positions staying at the same spot in both I(t) 
    and I(t-1), the output image will be blank and black.
    Background subtraction seems that anything that is in the image that was not present
    in the first input image aka the background, will be hightlighted. It is useful for
    surveillance to detect objects that were not present in the first background image. 
```
