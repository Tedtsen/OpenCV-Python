```python
#double threshold for color range detection

import cv2
import numpy as np

def MaxCallback(pos):
    pass
def MinCallback(pos):
    pass

#Create Trackbars
cv2.namedWindow('Test')
cv2.createTrackbar('HueMin', 'Test',   60, 179, MinCallback)
cv2.createTrackbar('HueMax', 'Test',  140, 179, MaxCallback) 
cv2.createTrackbar('SatMin', 'Test',  10, 255, MinCallback)
cv2.createTrackbar('SatMax', 'Test', 150, 255, MaxCallback) 
cv2.createTrackbar('ValMin', 'Test', 220, 255, MinCallback)
cv2.createTrackbar('ValMax', 'Test', 255, 255, MaxCallback) 

out_size = (480, 270)
cap = cv2.VideoCapture('WiiPlay.mp4')

#Cut video length
time_length = 1481.0
fps=30
frame_seq = (9*60+3)*30
no_frame_processed = 0
frame_limit = (9*60+11)*30 - frame_seq
cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq);

# Check if the video file is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open the video file")
 
while True:
    ret, frame = cap.read()
    no_frame_processed += 1
    if ret == False:
        break
    if no_frame_processed == frame_limit:
        break
    cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)
    cv2.imshow('Input', cur_frame)
    
    #Convert BGR to HSV
    hsv = cv2.cvtColor(cur_frame, cv2.COLOR_BGR2HSV)
    
    #Get trackbars values
    HueMin = cv2.getTrackbarPos('HueMin','Test')
    HueMax = cv2.getTrackbarPos('HueMax','Test') 
    SatMin = cv2.getTrackbarPos('SatMin','Test')
    SatMax = cv2.getTrackbarPos('SatMax','Test') 
    ValMin = cv2.getTrackbarPos('ValMin','Test')
    ValMax = cv2.getTrackbarPos('ValMax','Test') 
    
    lower_colour = np.array([HueMin, SatMin, ValMin])
    upper_colour = np.array([HueMax, SatMax, ValMax])
    mask = cv2.inRange(hsv, lower_colour, upper_colour)
    
    #Morphological filters (Opening)
    kernel = np.ones((2,1), np.uint8)
    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    #Blue Cursor Detection Threshold
    cv2.namedWindow('Cursor')
    cursor = cv2.inRange(hsv, np.array([60, 10, 255]), np.array([140, 150, 255]))
    cv2.imshow('Cursor', cursor)
    
    #Yellow Timer Detection Threshold
    cv2.namedWindow('Timer')
    timer = cv2.inRange(hsv, np.array([0, 200, 200]), np.array([50, 255, 255]))
    cv2.imshow('Timer', timer)
    
    #Human Skin Detection Threshold
    cv2.namedWindow('Skin')
    skin = cv2.inRange(hsv, np.array([0, 50, 60]), np.array([25, 180, 225]))
    cv2.imshow('Skin', skin)
    
    cv2.imshow('Test', opening)
    
    c = cv2.waitKey(40)
    if c == 32:
        cv2.waitKey()
    if c == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
