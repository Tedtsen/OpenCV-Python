```python
#write video file

import cv2

cap = cv2.VideoCapture('WiiPlay.mp4')
# Check if the video file is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open the video file")
    
#BE CAREFUL! Each video has different time length and frame rate. 
#So make sure that you have the right parameters for the right video!
time_length = 1481.0
fps=30
frame_seq = (9*60+3)*30
out_size = (640, 360)
no_frame_processed = 0
frame_limit = (9*60+11)*30 - frame_seq

# 使用 X264 編碼
#fourcc = cv2.VideoWriter_fourcc('m','p','4','v')
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# 建立 VideoWriter 物件，輸出影片至 output.mp4
# FPS 值為 30.0，解析度為 640x360
out = cv2.VideoWriter('output.mp4', fourcc, 30.0, out_size)

#The first argument flag CAP_PROP_POS_FRAMES is a 0-based index of the frame to be decoded/captured next.
#The second argument defines the frame number in range from 0 to (time_length*fps - 1).
cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq);

ret, frame = cap.read()
cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)
while True:
    ret, frame = cap.read()
    no_frame_processed += 1
    if ret == False:
        break
    if no_frame_processed == frame_limit:
        break
    cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)
    ##################################
    height, width, channel = cur_frame.shape
    rgb_cur_frame = cv2.cvtColor(cur_frame, cv2.COLOR_BGR2RGB)
    for y in range(0, height):
        break_flag = 0
        for x in range(0, width):
            if rgb_cur_frame[y,x,2] >=180 and rgb_cur_frame[y,x, 0] <= 30 and rgb_cur_frame[y,x, 1] <= 60:
                target_x1 = x - 20
                target_y1 = y
                target_x2 = x + 40
                target_y2 = y + 60
                cv2.rectangle(rgb_cur_frame, (target_x1, target_y1), 
                            (target_x2, target_y2), (255,0,0), 2)
                break_flag = 1
                break
        if break_flag == 1:
            break
    cur_frame = cv2.cvtColor(rgb_cur_frame, cv2.COLOR_RGB2BGR)
    ##################################
    cv2.imshow('downsampling', cur_frame)
    out.write(cur_frame)
    c = cv2.waitKey(1)
    if c == 27:
        break
cap.release()
#must release the output 
out.release()
cv2.destroyAllWindows()
```
