```python
import cv2
import numpy as np

#nothing Definition
def nothing(x):
    pass
    
cap = cv2.VideoCapture('WiiPlay.mp4')

if not cap.isOpened():
    raise IOError("Cannot open the video file")
    
time_length = 1481.0
fps=30
frame_seq = (9*60+3)*30
out_size = (640, 360)
no_frame_processed = 0
frame_limit = (9*60+11)*30 - frame_seq


fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 30.0, out_size)

cap.set(cv2.CAP_PROP_POS_FRAMES , frame_seq);

ret, frame = cap.read()
cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)

#Create Trackbar Outside of Loop In Window Named 'Output'
#######################
cv2.namedWindow('Output')
cv2.createTrackbar('low', 'Output', 100, 179, nothing)
cv2.createTrackbar('high', 'Output', 140, 179, nothing)
#######################

while True:
    ret, frame = cap.read()
    no_frame_processed += 1
    if ret == False:
        break
    if no_frame_processed == frame_limit:
        break
    cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)
    cv2.imshow('original', cur_frame)
    #####################################
    #process cur_frame
    hsv = cv2.cvtColor(cur_frame, cv2.COLOR_BGR2HSV)
    #Get Trackbar Pos from Trackbars in Window Named 'Output'
    low = cv2.getTrackbarPos('low','Output')
    high = cv2.getTrackbarPos('high','Output') 
    
    lower_colour = np.array([low, 10, 255]) 
    upper_colour = np.array([high, 150, 255])
    mask = cv2.inRange(hsv, lower_colour, upper_colour)
    #Show Img In Window Named 'Output'
    cv2.imshow('Output', mask)
    #####################################
    out.write(cur_frame)
    c = cv2.waitKey(27)
    if c == 27:
        break
cap.release()
#must release the output 
out.release()
cv2.destroyAllWindows()
```
